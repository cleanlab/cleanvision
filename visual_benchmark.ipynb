# Analyze using cifar10
#### download the data

```python
# pip install cifar2png
# cifar2png cifar10 ./cifar10png --name-with-batch-index

categories = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]
data_types = ["train", "test"]
results_df = None

for data_type in data_types:
    for category in categories:
        DATA_DIR = f'./cifar10png/{data_type}/{category}/'
        print(f"Loading images from {DATA_DIR}")
        imagelab = Imagelab(DATA_DIR, thumbnail_size=(128, 128))
        print(f"Number of images: {str(len(imagelab.image_files))}")

        issue_types = ["Brightness"]
        # issue_types = None

        imagelab.find_issues(issue_types)
        imagelab.aggregate(thresholds=5)
        imagelab.summary()
        results_df = imagelab.results if results_df is None else pd.concat([results_df,imagelab.results])
results_df
```

# Analyze using caltech256
