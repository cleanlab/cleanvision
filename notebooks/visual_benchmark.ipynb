# Artificially inject image issues into caltech and see if we can identify them

```python
import os 
import random
import numpy as np 
from PIL import Image, ImageFilter 
import pandas as pd

DATA_DIR = '/Users/sanjana/Downloads/synthetic_caltech'
info_csv = pd.read_csv(f'{DATA_DIR}/caltech256.csv')
filenames = info_csv['filepath'].tolist()
filenames = [f'{DATA_DIR}/{file}' for file in filenames]
error_types = ["AspectRatio", "Blurred", "Duplicated", "Entropy", "Grayscale", "NearDuplicates"]

def blur_image(image):
    radius = np.random.choice(np.arange(2, 10))
    image = image.filter(ImageFilter.GaussianBlur(radius = radius))
    return image 

def change_aspect_ratio(image):
    height, width = image.size
    delta = np.random.choice([int(height*0.1), int(width*0.2)])
    height = height + (random.choice((-1,1)) * delta)
    width = width + (random.choice((-1,1)) * delta)

    image = image.resize((width, height), Image.Resampling.LANCZOS)
    return image

def get_near_duplicate(image):
    from PIL import ImageChops
    
    a = ImageChops.offset(image, -4, 0)
    b = blur_image(image)

    return np.random.choice([a,b])

def get_duplicate(image):
    return image 
    
def error_entropy(image):
    height, width = image.size
    color = np.random.choice(np.arange(256))
    new_height = np.random.choice([int(0.9*height), height])
    new_width = np.random.choice([int(0.8*width), width])
    
    image.paste((color, color, color), [0,0, new_height, new_width])
    return image

def change_to_gray(image):
    import PIL.ImageOps as ImageOps 
    return ImageOps.grayscale(image)

function_dict = {
    "AspectRatio": change_aspect_ratio, 
    "Blurred": blur_image, 
    "Duplicated": get_duplicate, 
    "Entropy": error_entropy, 
    "Grayscale": change_to_gray, 
    "NearDuplicates": get_near_duplicate
}

def induce_error(filenames, error_types = error_types, num_errors = 10):
    files_to_change = random.sample(filenames, len(error_types) * num_errors)
    
    index = 0
    files_to_delete = []
    for error_type in error_types:
        for _ in range(num_errors):
            filename, ext = os.path.splitext(os.path.basename(files_to_change[index]))
            image = Image.open(files_to_change[index])
            new_image = function_dict[error_type](image)
            new_filename = f"{filename}_{error_type}{ext}"
            savepath = f"{os.path.dirname(files_to_change[index])}/{new_filename}"
            new_image.save(savepath)
        
            if error_type != ["Duplicated", "NearDuplicates"]:
                files_to_delete.append(files_to_change[index])
        
            index += 1
    
    for file in files_to_delete:
        os.remove(file)

induce_error(filenames, num_errors=50)
```
